# Which categories, relating to consumer spending, have decreased in popularity since pandemic? #


In this data analysis notebook, I look at what goods and services have decreased in popularity since the pandemic. I specifically look at travel and workout classes. I used Google Trends data for this analysis in order to see the popularity. 

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## Travel ##

### Charts and Graphs that show the decrease in popularity of Marriott ###

```python
Marriott_df = pd.read_csv('../data/Google_Trends/Marriott.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
Marriott_df
```

```python
# plot the Google Trends ranking data with a line plot
# time series
Marriott_df.plot(figsize=(12,4))
```

This graph shows the decrease in popularity of Marriott hotels using Google Trends data. The popularity is ranked on the y-axis and the week is on the x-axis. There was a very steep decrease right when lockdown was put in place and it started to go back up again, but not to the pre-pandemic levels. It probably started to go back up because some lockdown orders were lifted and the hotels had a lot of cleaning procedures that made people feel more safe at their hotels. 

## Workout Classes ##

### Charts and Graphs that show decrease in popularity in Workout Classes

```python
FitnessClasses_df = pd.read_csv('../data/Google_Trends/FitnessClasses.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
FitnessClasses_df
```

```python
# plot the Google Trends ranking data with a line plot
# time series
FitnessClasses_df.plot(figsize=(12,4))
```

This graph shows the decreased popularity of searching for fitness classes using Google Trends data. It appears as though it has remained at a low level since the pandemic without any spikes back to normalcy. This is probably due to shut downs and consumers not feeling comfortable going to workout classes. Additionally, people probably got a routine down with an at-home workout service and do not need to risk getting sick by going somewhere.
