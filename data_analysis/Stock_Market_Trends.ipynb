# What does stock market trends and company performance tell us about consumer prefrences since Covid-19? #

In this data analysis notebook, I am looking at companies that have been performing well and poorly since the start of the pandemic. This allows me to see what consumers have been preferring by looking at the performance of various companies. I am looking at Peloton, Clorox. Netflix, Zoom, Amazon, Ralph Lauren, Delta Airlines, and Carnival Cruise. I used Yahoo Finance data for this data.

## Companies That Have Been Performing Well##

### Peloton Stock ###

```python
PTON = pd.read_csv('../data/Stock_Market_Data/PTON.csv')
```

```python
# 1. set a datetime index
PTON=PTON.set_index(pd.to_datetime(PTON['Date']))

# 2. select column to plot
PTON['Close'].plot()
```

```python
PTON.resample('M')['Close'].mean().plot(kind='bar')
```

__Observations:__
This data, which is from Yahoo Finance, shows that Peloton stock and the company in general has been doing well since the pandemic because it is a service that allows you to workout from home. Not only do they have bikes and treadmills that you can take classes on, but if one does not want to purchase an actual piece of workout equipment, you could buy the monthly subscription to the classes. This allows you to do many different types of workout classes from your own home and with your own equipment. This business model has proved to be successful since the pandemic hit, and the company is gaining a lot of popularity. 

In terms of consumer preferences, people want goods and services that allow them to feel safe and stay at home. Therefore, peloton's services have been very popular.
