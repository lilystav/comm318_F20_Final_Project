# What does comparing similar services tell us about consumer prefrences?#

In this data analysis, I compare similar services to see which are gaining popularity. I specifically look at services for watching movies or television and services for travel. I used Google Trends data for this analysis in order to compare the popularity. 

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## Comparing services for watching movies##

__This data shows the difference in popularity between Disney+, on online streaming service, and AMC Theaters__

```python
Streaming_df = pd.read_csv('../data/Google_Trends/Streaming.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
Streaming_df
```

```python
Streaming_df.shape
```

```python
# plot the Google Trends ranking data with a line plot
# time series
Streaming_df.plot(figsize=(12,4))
```

This graph uses Google Trends data and shows that there was an increase in popularity of Disney+ after March, when lockdown was in place. It is interesting to compare this with a decrease in popularity of AMC Theaters, due to closures and restrictions put in place from the pandemic. Also, you can see that in the beginning of the year, there was an increase in Disney+ popularity as well. This could be due to the fact it was a new service and people were interested in it. This overall shift in popularity is due to consumer spending becoming more digital. 

## Comparing services for travel ##

__This data shows the difference in popularity between Airbnb, a service to rent homes around the world, and Expedia, an online travel shopping company.__ 

```python
Travel_df = pd.read_csv('../data/Google_Trends/Travel.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
Travel_df
```

```python
Travel_df.shape
```

```python
# plot the Google Trends ranking data with a line plot
# time series
Travel_df.plot(figsize=(12,4))
```

This comparison shows the difference in popularity between Expedia and Airbnb using Google Trends data. They are both websites for travel, but it is clear that Expedia has become less popularity while Airbnb has become more popular. They had about the same level of popularity before the pandemic hit. This discrepancy in popularity between these two services could be because Airbnb has been better able to adapt to people wanting to vacation in a place close to their homes. It allows people to travel, and stay in a house or apartment anywhere they want which is useful now because people want to go somewhere they can drive. Where Expedia is a different service in that it plans your vacation, and usually involves flying and staying in hotels. Many people might not be comfortable with air travel and staying in hotels so it makes sense that Expedia has decreased in popularity during these times. This tells us that people are spending their money, in this instance when it comes to travel, in ways where they can still feel safe and do not have to travel too far.
