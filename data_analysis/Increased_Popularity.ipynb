# Which categories, relating to consumer spending, have increased in popularity since pandemic? #

In this data analysis notebook, I look at what goods and services have increased in popularity since the start of the pandemic. I am looking at E-commerce necessary goods, and at home services. I used Google Trends data for this analysis in order to see the popularity. 

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## E-Commerce ##

### Chart and Graph that show increase in popularity of online grocery services

__Amazon Fresh Data__

```python
fresh_df = pd.read_csv('../data/Google_Trends/AmazonFresh.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
fresh_df
```

```python
fresh_df.shape
```

There are 52 rows of data and each of them represent a week in the year. The column on the left is the week and the column on the right is the popularity (ranked from 1 to 100). 

```python
# plot the Google Trends ranking data with a line plot
# time series
fresh_df.plot(figsize=(12,4))
```

This graph shows how there was a big increase in popularity of Amazon Fresh using Google Trends data. This is not limited to Amazon Fresh, many  grocery delivery services increased in popularity as well. This is due to people being scared to go to grocery stores and leave their houses during the pandemic. It is interesting to see that this large increase in popularity did not remain constant. It was probably because people regained confidence in going to a grocery store or were tired of being home. There are also more increases in October and November, probably due to cases rising again or people realizing the convenience of such services and want to continue using them.
